## Linear Regression
$
\begin{align*}
y_i &\sim{} \mathcal N\left(\mathbf{x}_i^\top \boldsymbol \beta, \sigma^2\right) && i = 1,\cdots,n\\
 &\;\quad {} \textrm{Assumptions} \\
1. &\;\quad {} \mathbf{x}_i^\top \boldsymbol \beta && \textrm{Linear model form true} && \hat y_i = \mathbf{x}_i^\top \boldsymbol{\hat \beta} \textrm{ from least squares}&& \hat \beta = \underset{\beta}{\textrm{argmin}} \sum_{i=1}^n(y_i - \hat y_i)^2 = \underset{\beta}{\textrm{argmin}} \; \hat{\textbf{e}}^\top\hat{\textbf{e}} \\
2. &\;\quad {} \sigma^2 && \textrm{Homoskedastity holds} && \sigma^2\neq\sigma_{x_i}^2 \textrm{ a function of } \mathbf{x}_i\\
3. &\;\quad {} \mathbf{x}_i && \textrm{Perfectly measured $\mathbf{x}_i$} && \mathbf{x}_i \textrm{ has no "uncertainty" }&&(\textit{$\mathbf{x}_i$ has no distributional assumptions})\\
4. &\;\quad {} \epsilon_i=y_i-\mathbf{x}_i^\top \boldsymbol \beta && \textrm{Normality of error terms} && \textrm{and residuals } \hat e_i=y_i-\hat y_i \\
\end{align*}$


## Generative Modeling

$p(y_i | \mathbf{x}_i, \boldsymbol {\hat \beta}, \hat \sigma)$ estimated as $\mathcal N\left(\mathbf{x}_i^\top \boldsymbol {\hat \beta}, \hat \sigma^2\right)$ is a **predictive model** as opposed to a **generative model**

Integrating out the estimation uncertainty provides the **posterior predictive distribution**

$p(y_i | \mathbf{x}_i) = \int p(y_i | \mathbf{x}_i, \boldsymbol {\hat \beta}, \hat \sigma) \; p(\boldsymbol {\hat \beta}, \hat \sigma)\; d \boldsymbol {\hat \beta} \; d \hat \sigma $

**Generative modeling** means considering stochastic rather than fixed $\mathbf{x}_i$

$p(y_i, \mathbf{x}_i) = p(y_i | \mathbf{x}_i) p(\mathbf{x}_i)  = \int p(y_i | \mathbf{x}_i, \boldsymbol \theta) \; p(\mathbf{x}_i | \boldsymbol \theta) \; p(\boldsymbol \theta)\; d \boldsymbol {\theta}$

## Least Squares

$
\begin{align*}
\hat \beta = \underset{\beta}{\textrm{argmin}} \sum_{i=1}^n(y_i - \hat y_i)^2 &={}
\underset{\beta}{\textrm{argmin}} \; \hat{\textbf{e}}^\top\hat{\textbf{e}}\\
&={} \underset{\beta}{\textrm{argmin}} \; (\textbf{y}-\textbf{X}\boldsymbol \beta)^\top(\textbf{y}-\textbf{X}\boldsymbol \beta)\\
&={} \underset{\beta}{\textrm{argmin}} \; \textbf{y}^\top \textbf{y} -2 (\textbf{X}\boldsymbol \beta)^\top \textbf{y} + (\textbf{X}\boldsymbol \beta)^\top (\textbf{X}\boldsymbol \beta)\\
\\
0 &={} \nabla_\beta \left[ \textbf{y}^\top \textbf{y} -2 \boldsymbol \beta^\top \textbf{X}^\top \textbf{y} + \boldsymbol \beta^\top \textbf{X}^\top \textbf{X}\boldsymbol \beta \right]\\
 &={} -2  \textbf{X}^\top \textbf{y} + 2 \textbf{X}^\top \textbf{X}\boldsymbol {\hat \beta} \\
 &={} - \textbf{X}^\top \textbf{y} + \textbf{X}^\top \textbf{X}\boldsymbol {\hat \beta} \\
 \\
\textbf{X}^\top \textbf{X}\boldsymbol {\hat \beta} &={} \textbf{X}^\top \textbf{y}  \\
\boldsymbol {\hat \beta} &={} (\textbf{X}^\top \textbf{X})^{-1}\textbf{X}^\top \textbf{y}  \\
\hat{\textbf{y}} = \textbf{X} \boldsymbol {\hat \beta} &={} \textbf{X} (\textbf{X}^\top \textbf{X})^{-1}\textbf{X}^\top \textbf{y} = \textbf{H} \textbf{y} = \hat{\textbf{y}} \\
\\
\textbf{X}^\top (\textbf{y}-\hat{\textbf{y}}) &={} \textbf{X}^\top \textbf{y} - \textbf{X}^\top\textbf{X} (\textbf{X}^\top \textbf{X})^{-1}\textbf{X}^\top \textbf{y} = \textbf{X}^\top \textbf{y} - \textbf{X}^\top \textbf{y} = \textbf{0} \quad  \Longrightarrow{} \quad \sum_{i=1}^n \hat e_i = 0 \\
&\Longrightarrow{} \quad \textbf{X}^\top \textbf{y} =   \textbf{X}^\top \hat{\textbf{y}}
\end{align*}$

## Generalized Linear Models (GLM)

1. Not all data has **normally distributed error terms**
2. $E[Y|\textbf{x}_i] = \textbf{x}_i^T \boldsymbol \beta$ can paramaterize more than a **normal distribution**

$E[Y|\textbf{x}_i] = e^{\textbf{x}_i^T \boldsymbol \beta }$ parameterizes $\lambda > 0$ of a **Poisson distribution**

> A **Poisson** GLM uses a **log link function** so $\log(\lambda_i) = \textbf{x}_i^T \boldsymbol \beta$

$E[Y|\textbf{x}_i] = \frac{1}{1 + e^{-\textbf{x}_i^T \boldsymbol \beta }}$ parameterizes $p \in [0,1]$ of a **Bernoulli distribution**

> A **logistic regression** uses a **logit (log odds) link function** so $\frac{p}{1-p_i} = \textbf{x}_i^T \boldsymbol \beta$

An alternative is the **probit link function** $\Phi^{-1}(p_i) = \textbf{x}_i^T \boldsymbol \beta$ where $\Phi^{-1}$ is the inverse CDF of a standard normal distribution; thus, $E[y|\textbf{x}_i] = p_i = \Phi(\textbf{x}_i^T \boldsymbol \beta) = Pr(z_i \geq 0) \; \textrm{ for } \; z_i \sim \mathcal N(\textbf{x}_i^T \boldsymbol \beta,1)$

$
\begin{align*}
\Phi(\textbf{x}_i^T \boldsymbol \beta)\sqrt{2\pi} &={} \int_{-\infty}^{x_i^\top\boldsymbol{\beta}_{p\times 1}} e^{-\frac{1}{2}z^2_i}dz\\
&={} \int_{-\infty}^{0} e^{-\frac{1}{2}(z_i + x_i^\top\boldsymbol{\beta}_{p\times 1})^2}dz = \int_{-\infty}^{0} e^{-\frac{1}{2}(-z_i - x_i^\top\boldsymbol{\beta}_{p\times 1})^2}dz =  \int^{\infty}_{0} e^{-\frac{1}{2}(z_i-x_i^\top\boldsymbol{\beta}_{p\times 1})^2}dz
\end{align*}$

This formulation maps **binary** $y_i$ to a **latent normally distributed representation** $z_i$. This demonstrates a fundamental idea underlying **copulas** which transform arbitrary distribution into latent normal distributions (several of which) which may then be tied together as a multivariate normal distribution to provide inference on an underlying **covariance structure**.

- The general **copula** formulation for $j^{th}$ random variable $\textbf{Y}_j$ with CDF $F_{Y_j}$ is $\Phi^{-1}(F_{Y_j}(y_{ij}))$ but the trick used above is to instead map this as $\Phi^{-1}(F_{Y_j}(E[Y_{ij}|\textbf{x}_i|))$

## Shrinkage Regularization

**Machine Learning** fits models by optimizing penalized **loss functions**.

Two classic regularizations are "ridge" and "lasso" regression, which respectively use $L_2$ and $L_1$ penalty functions.

- Ridge: $$\sum_{i=1}^n \frac{1}{2}(y_i-x_i^T\beta_{p \times 1})^2 + \lambda \sum_{j=1}^n \beta_j^2 = \frac{1}{2}(y-X\beta)^T(y-X\beta) + \lambda \sum_{j=1}^n \beta_j^2 = \frac{1}{2}||y-X\beta||_2^2 + \lambda ||\beta||_2^2 $$
- Lasso: $$\sum_{i=1}^n \frac{1}{2}(y-x_i^T\beta_{p \times 1})^2 + \lambda \sum_{j=1}^n |\beta_j| = \frac{1}{2}(y-X\beta)^T(y-X\beta) + \lambda \sum_{j=1}^n |\beta_j| = \frac{1}{2}||y-X\beta||_2^2 + \lambda ||\beta||_1$$

### As priors

Bayes has been regularizing since [1763](https://en.wikipedia.org/wiki/Thomas_Bayes#Bayes'_theorem)

- **Machine Learning** is the original inventor or "regularization"?
- Bayesian analysis has been regularizing since it's inception: *priors are regularization*


**Ridge Regression** **$L_2$-penalty** is just a MAP estimate using **norml priors** for the **regression coefficients**

\begin{align*}
 \beta_i \sim {} & \text{Normal}(b_i,s_i) & f(\beta_i| b_i, s_i) = {}& \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\left(\frac{\beta_i-b_i}{s_i}\right)^2} & \overbrace{\underbrace{|\beta_i-b_i|^2}}_{\text{Squared Penalization}}^{L_2} \\
y_i \sim {} & \text{Normal}(x_i^T\beta, \sigma) & \sigma \sim {} & \text{HalfNormal}(\sigma_0)
\end{align*}

**Lasso Regression** **$L_1$-penalty** is just replacing the **normal priors** with **Laplace ("Double Exponential") prior distributions**

\begin{align*}
\beta_i \sim {} & \text{Laplace}(b_i,s_i) & f(\beta_i| b_i, s_i) = {}& {\frac  {1}{2b}}\exp \left(-{\frac{|\beta_i-b_i |}{s_i}}\right) & \overbrace{\underbrace{|\beta_i-b_i|}}_{\text{Absolute Penalization}}^{L_1}\\
y_i \sim {} & \text{Normal}(x_i^T\beta, \sigma) & \sigma \sim {} & \text{HalfNormal}(\sigma_0)
\end{align*}

### **Q1: derive the following**

1. $p(\boldsymbol \beta | \textbf{y}, \textbf{X}, \Sigma=\sigma^2 I)$ for the **linear regression model** likelihood proportional to $\exp \left(-{\frac {1}{2}}(\mathbf {y} -\mathbf{X\boldsymbol \beta})^\top \Sigma^{-1}(\mathbf {y} - \mathbf{X\boldsymbol\beta})\right)$ and a $\mathcal{MVN}(\boldsymbol\beta_0, \boldsymbol\Sigma_\beta)$ for $\boldsymbol\beta$

2. The analytical posterior distribution for $\sigma^2$ the **error variance** of a **linear regression model** with **design matrix** $\mathbf{X}$ assuming $\sigma^2$ has an **inverse-gamma** prior distribution with parameters $\alpha^*$ and $\beta^*$ (unrelated to $\boldsymbol \beta$).

### **Solution to Q1**

### 1. Derivation of  
$$
p(\boldsymbol \beta \mid \mathbf{y}, \mathbf{X}, \Sigma=\sigma^2I)
$$  
when the likelihood is
$$
p(\mathbf{y}\mid \boldsymbol\beta, \sigma^2, \mathbf{X}) \propto \exp\!\Biggl(-\frac{1}{2}(\mathbf{y}-\mathbf{X}\boldsymbol\beta)^\top \Bigl(\frac{1}{\sigma^2}I\Bigr)(\mathbf{y}-\mathbf{X}\boldsymbol\beta)\Biggr)
=\exp\!\Biggl(-\frac{1}{2\sigma^2}(\mathbf{y}-\mathbf{X}\boldsymbol\beta)^\top(\mathbf{y}-\mathbf{X}\boldsymbol\beta)\Biggr)
$$
and the prior is
$$
p(\boldsymbol \beta) \propto \exp\!\Biggl(-\frac{1}{2}(\boldsymbol\beta-\boldsymbol\beta_0)^\top \boldsymbol\Sigma_\beta^{-1} (\boldsymbol\beta-\boldsymbol\beta_0)\Biggr).
$$

**Step 1. Write the joint (unnormalized) posterior:**  
Multiplying likelihood and prior (ignoring normalizing constants) yields
$$
p(\boldsymbol\beta \mid \mathbf{y}, \mathbf{X}, \sigma^2) \propto \exp\!\Biggl\{-\frac{1}{2\sigma^2}(\mathbf{y}-\mathbf{X}\boldsymbol\beta)^\top(\mathbf{y}-\mathbf{X}\boldsymbol\beta)
-\frac{1}{2}(\boldsymbol\beta-\boldsymbol\beta_0)^\top\boldsymbol\Sigma_\beta^{-1}(\boldsymbol\beta-\boldsymbol\beta_0)\Biggr\}.
$$

**Step 2. Expand the quadratic form in $\boldsymbol\beta$:**  
The likelihood part expands as
$$
\frac{1}{2\sigma^2}(\mathbf{y}-\mathbf{X}\boldsymbol\beta)^\top(\mathbf{y}-\mathbf{X}\boldsymbol\beta)
=\frac{1}{2\sigma^2}\Bigl[\mathbf{y}^\top\mathbf{y}-2\,\mathbf{y}^\top\mathbf{X}\boldsymbol\beta+\boldsymbol\beta^\top\mathbf{X}^\top\mathbf{X}\boldsymbol\beta\Bigr].
$$
The term $\mathbf{y}^\top\mathbf{y}$ does not depend on $\boldsymbol\beta$ and can be absorbed in the constant.

Thus, combining with the prior, the exponent (ignoring additive constants) becomes
$$
-\frac{1}{2}\Biggl\{\boldsymbol\beta^\top\Bigl(\frac{1}{\sigma^2}\mathbf{X}^\top\mathbf{X} + \boldsymbol\Sigma_\beta^{-1}\Bigr)\boldsymbol\beta
-2\,\boldsymbol\beta^\top\Bigl(\frac{1}{\sigma^2}\mathbf{X}^\top\mathbf{y}+\boldsymbol\Sigma_\beta^{-1}\boldsymbol\beta_0\Bigr)\Biggr\}.
$$

**Step 3. Recognize the kernel of a multivariate normal and complete the square:**  
This quadratic form in $\boldsymbol\beta$ shows that the posterior is a multivariate normal distribution with precision (inverse covariance)
$$
\Lambda = \frac{1}{\sigma^2}\mathbf{X}^\top\mathbf{X} + \boldsymbol\Sigma_\beta^{-1},
$$
and mean given by
$$
\boldsymbol\mu_\beta = \Lambda^{-1}\Bigl(\frac{1}{\sigma^2}\mathbf{X}^\top\mathbf{y}+\boldsymbol\Sigma_\beta^{-1}\boldsymbol\beta_0\Bigr).
$$

Thus, the **posterior distribution** is
$$
\boldsymbol\beta\mid \mathbf{y}, \mathbf{X}, \sigma^2 \sim \mathcal{MVN}\!\Biggl(\Lambda^{-1}\Bigl(\frac{1}{\sigma^2}\mathbf{X}^\top\mathbf{y}+\boldsymbol\Sigma_\beta^{-1}\boldsymbol\beta_0\Bigr),\ \Lambda^{-1}\Biggr)
$$
or, equivalently,
$$
\boxed{
\boldsymbol\beta\mid \mathbf{y}, \mathbf{X}, \sigma^2 \sim \mathcal{MVN}\!\left(\left(\frac{1}{\sigma^2}\mathbf{X}^\top\mathbf{X}+\boldsymbol\Sigma_\beta^{-1}\right)^{-1}\left(\frac{1}{\sigma^2}\mathbf{X}^\top\mathbf{y}+\boldsymbol\Sigma_\beta^{-1}\boldsymbol\beta_0\right),\ \left(\frac{1}{\sigma^2}\mathbf{X}^\top\mathbf{X}+\boldsymbol\Sigma_\beta^{-1}\right)^{-1}\right).
}
$$

### 2. Derivation of the analytical posterior for $\sigma^2$  
Assume the model is the same as above (with likelihood)
$$
p(\mathbf{y}\mid \boldsymbol\beta, \sigma^2, \mathbf{X}) \propto (\sigma^2)^{-n/2}\exp\!\left\{-\frac{1}{2\sigma^2}(\mathbf{y}-\mathbf{X}\boldsymbol\beta)^\top(\mathbf{y}-\mathbf{X}\boldsymbol\beta)\right\},
$$
and the prior for $\sigma^2$ is an inverse-gamma distribution:
$$
p(\sigma^2) \propto (\sigma^2)^{-\alpha^*-1}\exp\!\left\{-\frac{\beta^*}{\sigma^2}\right\}.
$$

**Step 1. Write the (unnormalized) full conditional for $\sigma^2$:**  
Multiplying the likelihood (conditional on $\boldsymbol\beta$) and the prior gives
$$
p(\sigma^2\mid \mathbf{y}, \mathbf{X}, \boldsymbol\beta) \propto (\sigma^2)^{-n/2} \exp\!\left\{-\frac{1}{2\sigma^2}S(\boldsymbol\beta)\right\} \cdot (\sigma^2)^{-\alpha^*-1}\exp\!\left\{-\frac{\beta^*}{\sigma^2}\right\},
$$
where
$$
S(\boldsymbol\beta) = (\mathbf{y}-\mathbf{X}\boldsymbol\beta)^\top(\mathbf{y}-\mathbf{X}\boldsymbol\beta).
$$

**Step 2. Combine the powers of $\sigma^2$ and the exponentials:**  
Collecting the terms gives
$$
p(\sigma^2\mid \cdot) \propto (\sigma^2)^{- \left(\alpha^* + \frac{n}{2} + 1\right)} \exp\!\left\{-\frac{1}{\sigma^2}\left(\frac{1}{2}S(\boldsymbol\beta) + \beta^*\right)\right\}.
$$

**Step 3. Recognize the kernel of an inverse-gamma:**  
The standard form for an inverse-gamma density is
$$
\text{IG}(\sigma^2; \alpha', \beta') \propto (\sigma^2)^{-\alpha'-1}\exp\!\left\{-\frac{\beta'}{\sigma^2}\right\}.
$$
By matching terms, we identify:
- Updated shape parameter:
  $$
  \alpha' = \alpha^* + \frac{n}{2},
  $$
- Updated scale parameter:
  $$
  \beta' = \beta^* + \frac{1}{2}S(\boldsymbol\beta) = \beta^* + \frac{1}{2}(\mathbf{y}-\mathbf{X}\boldsymbol\beta)^\top(\mathbf{y}-\mathbf{X}\boldsymbol\beta).
  $$

Thus, the **posterior for $\sigma^2$** (conditional on $\boldsymbol\beta$) is
$$
\boxed{
\sigma^2\mid \mathbf{y}, \mathbf{X}, \boldsymbol\beta \sim \text{Inverse-Gamma}\!\Bigl(\alpha^*+\frac{n}{2},\ \beta^*+\frac{1}{2}(\mathbf{y}-\mathbf{X}\boldsymbol\beta)^\top(\mathbf{y}-\mathbf{X}\boldsymbol\beta)\Bigr).
}
$$

---

### Final Answer Summary

1. **Posterior for $\boldsymbol\beta$:**
   $$
   \boldsymbol\beta\mid \mathbf{y}, \mathbf{X}, \sigma^2 \sim \mathcal{MVN}\!\left(\left(\frac{1}{\sigma^2}\mathbf{X}^\top\mathbf{X}+\boldsymbol\Sigma_\beta^{-1}\right)^{-1}\left(\frac{1}{\sigma^2}\mathbf{X}^\top\mathbf{y}+\boldsymbol\Sigma_\beta^{-1}\boldsymbol\beta_0\right),\ \left(\frac{1}{\sigma^2}\mathbf{X}^\top\mathbf{X}+\boldsymbol\Sigma_\beta^{-1}\right)^{-1}\right).
   $$

2. **Posterior for $\sigma^2$:**
   $$
   \sigma^2\mid \mathbf{y}, \mathbf{X}, \boldsymbol\beta \sim \text{Inverse-Gamma}\!\left(\alpha^*+\frac{n}{2},\ \beta^*+\frac{1}{2}(\mathbf{y}-\mathbf{X}\boldsymbol\beta)^\top(\mathbf{y}-\mathbf{X}\boldsymbol\beta)\right).
   $$

These derivations use standard conjugate results for Bayesian linear regression and ensure that the complete (conditional) posteriors for both $\boldsymbol\beta$ and $\sigma^2$ are correctly specified.

### **Q3: perform robust Bayesian Linear Regression**

Let $p(\tau_i)$ be $\require{cancel} \textrm{gamma}\big(\tau_i | \alpha = \frac{\nu}{2}, \overset{\textrm{rate}\xcancel{\textrm{scale}}}{\beta = \frac{\nu}{2}}\big)$ and let $p(y_i|\boldsymbol \beta, \tau,\tau_i)$ be $\mathcal{N}(y_i | \textbf{X} \boldsymbol \beta, \overset{\textrm{precision}}{\tau \times \tau_i})$.

- Return to your dataset and maniputate it to have some various outliers or find another dataset with some outlier data
- Use an appropriate prior for inference on $v$ if you have enough data to do so
- Use the posterior distributions of the $\tau_i$'s to identify data point "outliers"
- Use the posterior distributions of the $\sigma_i^{-2} = \tau \times \tau_i$ to create posterior distribuitions of the **influence** (the diagonals of the $H$ "hat" matrix $X^\top (X^\top D X)^{-1} X$ where $D_{ij}=0$ and $D_{ii} = \sigma^2_i$) and compare and contras some example "outlier" versus "non outlier" data points

- Provide inference with Bayesian posterior analysis and report MCMC diagnostics

### **Solution to Q3**

```python
import pymc as pm
import numpy as np
import arviz as az
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler

# ------------------------
# Data Preparation
# ------------------------
# Load the diabetes dataset
X, y = load_diabetes(return_X_y=True)
n, p = X.shape  # n should be 442 for the diabetes dataset
print("Original X shape:", X.shape)
print("Original y shape:", y.shape)

# Introduce outliers: randomly select a few indices and add a large offset to y
rng = np.random.default_rng(42)
n_outliers = 10
outlier_idx = rng.choice(n, size=n_outliers, replace=False)
y_out = y.astype(float).copy()
y_out[outlier_idx] += 3 * y.std()

# Standardize predictors and response for numerical stability
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X)
y_standardized = (y_out - y_out.mean()) / y_out.std()
print("X_scaled shape:", X_scaled.shape)
print("y_standardized shape:", y_standardized.shape)

# ------------------------
# Build the Robust Bayesian Regression Model
# ------------------------
with pm.Model() as robust_model:
    # Regression coefficients with independent Normal priors
    betas = pm.Normal('betas', mu=0, sigma=1, shape=p)

    # Global error scale: we put a HalfNormal prior on sigma and define tau = 1/sigma**2
    sigma = pm.HalfNormal('sigma', sigma=1)
    tau = pm.Deterministic('tau', 1/sigma**2)

    # Degrees-of-freedom parameter for the heavy tails (v or nu)
    nu = pm.Exponential('nu', lam=1/30)

    # Local precision multipliers (one per data point); shape must be n (442)
    tau_i = pm.Gamma('tau_i', alpha=nu/2, beta=nu/2, shape=n)

    # Likelihood: note that the observation-specific standard deviation is 1/sqrt(tau * tau_i)
    mu = pm.math.dot(X_scaled, betas)
    y_obs = pm.Normal('y', mu=mu, sigma=1/pm.math.sqrt(tau * tau_i), observed=y_standardized)

    # Sample from the posterior
    trace_robust = pm.sample(3000, tune=1500, target_accept=0.95, return_inferencedata=True)

# ------------------------
# MCMC Diagnostics
# ------------------------
print(az.summary(trace_robust, round_to=2))
az.plot_trace(trace_robust, var_names=['betas', 'sigma', 'nu'])
plt.show()

# Check the shape of tau_i in the posterior: should be (chain, draw, n)
print("Posterior tau_i shape:", trace_robust.posterior['tau_i'].shape)

# ------------------------
# Outlier Identification via the Local Precision Multipliers
# ------------------------
# Compute the posterior mean of tau_i across chains and draws (should be length n)
tau_i_means = trace_robust.posterior['tau_i'].mean(dim=("chain", "draw")).values

plt.figure(figsize=(8, 4))
plt.stem(np.arange(n), tau_i_means, basefmt=" ")
plt.xlabel('Observation index')
plt.ylabel('Posterior mean of tau_i')
plt.title('Posterior Means of Local Precision Multipliers (τᵢ)')
plt.axhline(np.percentile(tau_i_means, 25), color='red', linestyle='--', label='25th Percentile')
plt.legend()
plt.show()

flagged_indices = np.where(tau_i_means < np.percentile(tau_i_means, 25))[0]
print("Flagged potential outlier indices (τᵢ below 25th percentile):", flagged_indices)

# ------------------------
# Compute Posterior Influence via the Weighted Hat Matrix
# ------------------------
# Use a subset of posterior samples for speed
n_samples = 100  # number of posterior samples to use
# Stack chain and draw dimensions into one sample dimension
posterior_samples = trace_robust.posterior.stack(sample=("chain", "draw")).isel(sample=slice(0, n_samples))

# Initialize an array to store influence (hat values) for each observation and sample
hat_values = np.zeros((n_samples, n))

# For each posterior sample, compute the weighted hat (influence) values.
for i in range(n_samples):
    # Extract the posterior sample for tau and tau_i
    tau_sample = posterior_samples['tau'].values[i]
    tau_i_sample = posterior_samples['tau_i'].values[i]  # Removed the extra indexing with ":"

    # Verify the shape of tau_i_sample
    if tau_i_sample.shape[0] != n:
        print(f"Warning: tau_i_sample shape is {tau_i_sample.shape}, expected ({n},)")
        # If shape is incorrect, try to reshape it correctly
        if tau_i_sample.size == n:
            tau_i_sample = tau_i_sample.reshape(n)
        else:
            # If reshaping is not possible, skip this sample
            print(f"Skipping sample {i} due to incorrect shape")
            continue

    # Create the diagonal weight matrix: W = diag(tau_sample * tau_i_sample)
    W = np.diag(tau_sample * tau_i_sample)

    # Compute X^T W X
    XTWX = X_scaled.T @ W @ X_scaled
    inv_XTWX = np.linalg.inv(XTWX)

    # Compute influence for each observation: H_i = (x_i^T (X^T W X)^{-1} x_i) * (tau_sample * tau_i_sample[i])
    for j in range(n):
        xj = X_scaled[j, :].reshape(1, -1)
        hat_values[i, j] = (xj @ inv_XTWX @ xj.T)[0, 0] * (tau_sample * tau_i_sample[j])

# Compute the posterior mean hat value for each observation
hat_means = hat_values.mean(axis=0)

plt.figure(figsize=(8, 4))
plt.stem(np.arange(n), hat_means, basefmt=" ")
plt.xlabel('Observation index')
plt.ylabel('Posterior Mean Influence (Hat Value)')
plt.title('Posterior Influence (Weighted Hat Values)')
plt.show()

# Compare influence for flagged outliers vs. non-outliers:
flagged = flagged_indices
non_flagged = np.setdiff1d(np.arange(n), flagged_indices)

print("Mean hat value for flagged (potential outlier) observations:")
print(hat_means[flagged])
print("Mean hat value for non-flagged observations:")
print(hat_means[non_flagged])
```
